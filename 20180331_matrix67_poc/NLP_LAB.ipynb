單純做概念上的實作，沒有考慮演算法的效能：

* 生出 ngram
* 找出 term 在文章中的 `前` `後` 字元 (為了讓頭尾詞能計算，多加特殊字元 🦄🍬表示)
* 計算 entropy
* 簡單門檻過濾一下 `凝固程度` `自由度`

```python
def ngram(input_sentence, n = 2):
    ret = []
    for i in range(0, len(input_sentence) - n + 1):
        ret.append(input_sentence[i:i+n])
    return ret
```

```python
def gen_all_terms(max_gram, words):

    import re
    # 隨意加一下的 stopwords
    sentences = re.split('[\u3000，。（）〕〔／《》、」！「：\n\t \[\]\(\)a-zA-Z]', words)
    sentences = ["{}".format(x) for x in sentences if len(x) > 0]
    
    # 頭尾加個特殊字元，有些詞是很有機會用在開頭，或結尾的
    # 如果沒有人可它結合起，就沒有那個詞的 前一個 或 後一個字
    # 所以，我們取巧一下，在開頭加 獨角獸 結尾放 糖果
    sentences_padding = ["🦄{}🍬".format(x) for x in sentences if len(x) > 0]
    
    grams = []
    for n in range(2, max_gram + 1):
        for s in sentences:
            grams += ngram(s, n)
    grams = list(set(grams))
    
    terms = {}
    for g in grams:
        if g not in terms:
            # g 的 [左]，[右] 字集
            terms[g] = [[], []]
        for s in sentences_padding:
            try:
                for m in re.findall("(.)({})(.)".format(g), s):
                    left, word, right = m
                    if word not in terms:
                        continue
                    terms[word][0] += [left]
                    terms[word][1] += [right]
            except Error:
                print("error {}".format(g))
                continue
    return grams, terms
```

```python
from math import log
def entropy(terms):
    stat = {}
    for s in terms:
        if s not in stat:
            stat[s] = 1
        else:
            stat[s] = stat[s] + 1

    all_entropy = 0
    for x in stat:
        p = stat[x] / len(terms)
        all_entropy = all_entropy + (- p * log(p))

    return all_entropy
```

```python
def filter_terms(terms, 最小出現次數, 最小自由度, filter=True):

    print("最小出現次數 {}, 最小自由度 {}".format(最小出現次數, 最小自由度))

    for x in terms:
        if len(terms[x][0]) >= 最小出現次數:
            e = [entropy(terms[x][0]), entropy(terms[x][1])]
            # 凝固程度 或 自由度 都是 entropy，只是小的那一個原作者稱它為自由度。
            凝固程度 = max(e)
            自由度 = min(e)
            
            # 想 debug 時，就把 filter 關了，看所有結果唄
            if not filter:
                print(x, terms[x], "凝固程度: {:.2f}, 自由度 {:.2f}".format(凝固程度, 自由度))
                continue
            elif 自由度 > 最小自由度:
                print(x, terms[x], "凝固程度: {:.2f}, 自由度 {:.2f}".format(凝固程度, 自由度))
```
